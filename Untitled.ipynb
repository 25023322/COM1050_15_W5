### A3


```python
def so_nguyen_to(x):

    if x==2 or x==3 :
        return("True")
    if x>3:
        check=True
        for i in range(2,x):
            if x%i==0:
                check=False
        if check==False:
            return("False")
        else:
            return("True")
            
        
print(so_nguyen_to(56))
```

### A4


```python
def so_hoan_hao(x):
    uoc=[]
    for i in range(1,x):
        if x%i==0:
            uoc.append(i)
    if x==sum(uoc):
        return("True")
    else:
        return("False")
        
        
print(so_hoan_hao(28))
```
